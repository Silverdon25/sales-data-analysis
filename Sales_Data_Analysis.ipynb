This project analyses sales data to identify trends, top-performing products, and profitability insights.

```python
## 1. Data Loading
```

```python
import pandas as pd

df = pd.read_csv("Sample - Superstore.csv", encoding='latin1')

df.head()
```

## 2. Data Overview

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

df.groupby('Region')['Sales'].sum().plot(kind='bar')

plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')

plt.show()
```

### Insight
The West region generates the highest sales, indicating strong market demand. 
This suggests potential opportunities to further expand operations in this region.

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products

```

```python
top_products.plot(kind='barh')

import matplotlib.pyplot as plt
plt.title('Top 10 Products by Sales')
plt.xlabel('Sales')
plt.ylabel('Product')
plt.gca().invert_yaxis()

plt.show()
```

### Insight
A small number of products contribute significantly to total sales, indicating a concentration of revenue in key items.

```python
df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
```

```python
df.groupby('Region')['Profit'].sum().plot(kind='bar')

plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Profit')

plt.show()
```

### Insight
Some regions generate high sales but lower profit, suggesting inefficiencies such as high discounts or operational costs.

## Final Insights

- West region generates highest sales and profit
- A small number of products drive most revenue
- Some regions show lower profit despite strong sales, indicating cost inefficiencies

## Business Recommendation

- Focus marketing on high-performing regions (West, East)
- Review pricing/discount strategy in low-profit regions
- Promote top-selling products to maximise revenue
